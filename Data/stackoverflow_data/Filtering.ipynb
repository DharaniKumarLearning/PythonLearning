```python
import pandas as pd
```

```python
people = {
    "first" : ["Dharani", "Moulali", "Bharath"],
    "last" : ["Kumar", "Reddi", "Reddi"],
    "email" : ["gopavaram.dharani@gmail.com", "moulali.naguri@gmail.com", "reddybharath702@gmail.com"]
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['last'] == "Reddi"  # this returns a series of objects
```

```python
filt = (df['last'] == "Reddi")  # creating a filter variable
```

```python
df[filt] # passing the filt variable to dataframe
```

```python
df[(df['last'] == "Kumar")]  # We can directly pass the filter condition to the dataframe if we want
```

```python
df.loc[filt]  # we can use loc function to filter the data
```

```python
df.loc[filt, ["email", "last"]]
```

```python
filt = (df['last'] == "Reddi") & (df["first"] == "Moulali")  # having multiple filter condition
# & -- and, | -- or
```

```python
df.loc[filt]
```

```python
filt = (df['first'] == "Dharani") | (df['first'] == "Bharath")
```

```python
df.loc[filt]
```

```python
df.loc[~filt]  # to get the rows that doesn't match the filter condition
```

```python
high_sal = (survey_df["CompTotal"] > 7000000)
```

```python
survey_df.loc[high_sal,["DevType","NEWCollabToolsHaveWorkedWith","Country","CompTotal"]]
```

```python
countries = ["United States of America", "Afghanistan", "India"]
high_sal_countries_filt = (survey_df["CompTotal"] > 7000000) & (survey_df['Country'].isin(countries))
```

```python
pd.set_option("display.max_rows", 121)
survey_df.loc[high_sal_countries_filt,["DevType","NEWCollabToolsHaveWorkedWith","LanguageHaveWorkedWith","Country","CompTotal"]]
```
